## Proyecto Calidad del Aire en Colombia (EDA)

### Acerca del conjunto de datos

Esta base de datos es generada por El Instituto de Hidrología, Meteorología y Estudios Ambientales – (IDEAM), el cual son las autoridades ambientales obligadas a realizar el reporte de la información de calidad del aire o nivel de inmisión en SISAIRE. Nuestro conjunto de datos se compone de la siguiente información:

* ID Estacion: Identificador único asignado por el Subsistema de Información sobre Calidad del Aire - SISAIRE a cada estación de monitoreo.
* Autoridad Ambiental: Autoridad Ambiental encargada de operar el Sistema de Vigilancia de la Calidad del Aire en la jurisdicción donde se ubica la estación de monitoreo.
* Estación: Nombre específico que se designa a la estación de monitoreo de calidad del aire.
* Latitud: Coordenada geográfica que indica la latitud de la ubicación de la estación (Sistema de referencia espacial MAGNA- SIRGAS)
* Longitud: Coordenada geográfica que indica la longitud de la ubicación de la estación (Sistema de referencia espacial MAGNA- SIRGAS)
* Variable: Contaminante atmosférico o variable meteorológica que se mide en la estación de monitoreo de la calidad del aire, de acuerdo con las siguientes abreviaturas: 

    Contaminantes atmosféricos: 
    1. PM10: Material particulado menor a 10 micras. 
    2. PM2.5: Material particulado menor a 2.5 micras. 
    3. PST: Partículas Suspendidas Totales.
    4. NO: Óxido de Nitrógeno.
    5. NO2: Dióxido de Nitrógeno.
    6. SO2: Dióxido de Azufre.
    7. SO2_TR: Azufre Total Reducido.
    8. O3: Ozono troposférico.
    9. CO: Monóxido de Carbono.

    Variables meteorológicas: 
    1. VViento: Velocidad del Viento.
    2. DViento: Dirección del viento.
    3. Taire: Temperatura del aire.
    4. Taire2: Temperatura del aire a 2 metros.
    5. Taire10: Temperatura del aire a 10 metros.
    6. Pliquida: Precipitación Liquida.
    7. P: Presión atmosférica.
    8. HAire: Humedad del aire.
    9. HAire2: Humedad del aire a 2 m.
    10. HAire10: Humedad del aire a 10 m.
    11. RGlobal: Radiación Solar Global.
    12. RUVb: Radiación Ultravioleta.

* Unidades: Unidad de medida empleada para expresar el contaminante atmosférico o variable meteorológica
* Tiempo de exposición (horas): Período de referencia específico establecido normativamente (Resolución 2254/2017) para evaluar los niveles de inmisión o concentraciones de un contaminante atmosférico (solo aplica para contaminantes atmosféricos)
* Año: Año en el que se realizó la medición en la respectiva estación.
* Promedio: Promedio anual del nivel de inmisión o concentración a condiciones de referencia o estándar (std) del contaminante atmosférico o promedio anual del valor de la variable meteorológica.
* Suma: Total acumulado anual del nivel de inmisión o concentración a condiciones de referencia o estándar (std) del contaminante atmosférico o total acumulado anual del valor de la variable meteorológica.
* No. de datos: Cantidad total de datos validos recopilados durante el año de medición.
* Representatividad Temporal: Porcentaje del tiempo total de un período tiene datos válidos registrados
* Excedencias limite actual: Número total de veces en que el contaminante atmosférico ha superado el nivel máximo permisible vigente establecido para el respectivo tiempo de exposición durante el año de medición.
* Porcentaje excedencias limite actual: Proporción determinada por el número de veces en que el contaminante atmosférico ha superado el nivel máximo permisible vigente establecido para el respectivo tiempo de exposición, en relación con el número total de datos según el tiempo de exposición, durante el año de medición.
* Mediana: Mediana anual del nivel de inmisión o concentración a condiciones de referencia o estándar (std) del contaminante atmosférico o mediana anual del valor de la variable meteorológica
* Percentil 98: Medida estadística que indica el valor del contaminante atmosférico o de la variable meteorológica por debajo del cual se encuentra el 98% de los datos en el grupo durante el año de medición
percentil_98
* Máximo: Valor más alto registrado del contaminante atmosférico o de la variable meteorológica durante el año de medición.
* Fechas/horas del máximo: Fecha y hora en las que se registró el valor máximo durante el año de medición.
* Mínimo: Valor más bajo registrado del contaminante atmosférico o de la variable meteorológica durante el año de medición.
* Fechas/horas del mínimo: Fecha y hora en las que se registró el valor mínimo durante el año.
* Días de excedencias: Número total de días en que el contaminante atmosférico ha superado el nivel máximo permisible vigente establecido para el respectivo tiempo de exposición durante el año de medición.
* Código del Departamento: Código DANE asignado al departamento.
* Nombre del Departamento: Nombre del departamento
* Código del Municipio: Código del municipio
* Nombre del Municipio: Nombre del municipio
* Tipo de Estación: Ubicación de la estación (Sistema de referencia espacial MAGNA- SIRGAS)
* Ubicacion: ubicacion

### Exploración Básica de los Datos

```python
# Cargamos el conjunto de datos para previa exploración de ellos
import pandas as pd
calidad_aire = pd.read_csv('CALIDAD_DEL_AIRE_EN_COLOMBIA.csv')
calidad_aire.head()
```

Podemos notar que en nuestra base de datos contamos con 28 columnas, ya especificadas anteriormente. Para nuestro análisis de calidad del aire anual en Colombia, podriamos descartar algunas columnas que no aportan información clave sobre los contaminantes o que podrían ser redundantes. 

Columnas claves a mantener (importantes para el análisis):  
✅ Variable → Contaminante atmosférico o variable meteorológica.  
✅ Unidades → Para entender las mediciones.  
✅ Tiempo de exposición (horas) → Importante para evaluar los niveles de contaminación.  
✅ Año → Necesario para el análisis temporal.  
✅ Promedio → Indicador clave de la calidad del aire.  
✅ Suma → Puede ser útil para tendencias acumuladas.  
✅ No. de datos → Para evaluar la confiabilidad de los datos.  
✅ Representatividad Temporal → Para saber qué tan completos son los datos.  
✅ Excedencias límite actual → Indica cuántas veces se superó el límite permitido.  
✅ Porcentaje excedencias límite actual → Relacionado con la gravedad de la contaminación.  
✅ Mediana → Útil para análisis estadísticos.  
✅ Percentil 98 → Indica valores extremos de contaminación.  
✅ Máximo → Valor más alto registrado, importante para eventos críticos.  
✅ Mínimo → Valor más bajo registrado, útil para variabilidad.  
✅ Días de excedencias → Indica cuántos días fueron críticos.  
✅ Nombre del Departamento → Para análisis geográfico.  
✅ Nombre del Municipio → Para análisis más detallado por ciudad.  
✅ Tipo de Estación → Puede influir en los niveles de contaminación.  

Columnas que podemos descartar (menos relevantes para el análisis):  
❌ ID Estación → No aporta información útil para el análisis general.  
❌ Autoridad Ambiental → No influye directamente en los niveles de contaminación.  
❌ Estación → Nombre específico de la estación, pero no es clave para el análisis.  
❌ Latitud / Longitud → Si no necesitas un análisis geoespacial, puedes descartarlas.  
❌ Fechas/horas del máximo y mínimo → Si solo analizas datos anuales, no son tan relevantes.  
❌ Código del Departamento / Código del Municipio → Si ya tienes los nombres, los códigos pueden ser redundantes.  
❌ Ubicación → Parece redundante con otras columnas.  

Columnas que podemos usar mas adelante dependiendo del análisis:  
🔹 Análisis geoespacial, se usa Latitud y Longitud.  
🔹 Tendencias horarias, se usa Fechas/horas del máximo y mínimo.  






```python
# Mantenemos en un nuevo dataset, solo nuestras variables a usar o importantes
df = calidad_aire[['Variable', 'Unidades', 'Tiempo de exposición (horas)', 'Año', 'Promedio', 'Suma', 
                   'No. de datos', 'Representatividad Temporal', 'Excedencias limite actual', 
                   'Porcentaje excedencias limite actual', 'Mediana', 'Percentil 98', 'Máximo', 'Mínimo', 
                   'Días de excedencias', 'Nombre del Departamento', 'Nombre del Municipio', 
                   'Tipo de Estación']].copy()
df.head()
```

```python
# Editamos los nombres de nuestras columnas. 
# De tal manera que no tengan tildes o simbolos que puedan presentar problemas en el futuro

df = df.rename(columns={'Tiempo de exposición (horas)': 'Tiempo de exposicion (horas)',
                        'Año': 'Year',
                        'Máximo': 'Maximo', 
                        'Mínimo': 'Minimo', 
                        'Días de excedencias': 'Dias de excedencias', 
                        'Tipo de Estación': 'Tipo de Estacion'})
df.head()
```

```python
# Verificamos nuestros cambios en el nombre de columna y su tipo de dato
df.info()
```

```python
# Revisamos los valores unicos en nuestras columnas con el fin de revisar que no tengamos simbolos, 
# errores de escritura o espacios que nos afecten en el futuro

print('--- Valores unicos en el Nombre del Departamento ---')
print(df["Nombre del Departamento"].unique())
```

```python
print('--- Valores unicos en el Nombre del Municipio ---')
print(df["Nombre del Municipio"].unique())
```

```python
print('---Valores unicos en el Tipo de Estacion---')
print(df["Tipo de Estacion"].unique())
```

```python
# Eliminamos todas las tildes en nuestros datos

import unicodedata

# Función para remover tildes
def remover_tildes(texto):
    return "".join(
        c for c in unicodedata.normalize("NFD", texto) if unicodedata.category(c) != "Mn"
    )

# Aplicar la función a las columnas
df["Nombre del Departamento"] = df["Nombre del Departamento"].apply(remover_tildes).str.upper()
df["Nombre del Municipio"] = df["Nombre del Municipio"].apply(remover_tildes).str.upper()
```

```python
# Verificamos nuestros valores en el nombre del departamento
print(df["Nombre del Departamento"].unique())
```

```python
# Verificamos nuestros valores en el nombre del municipio
print(df["Nombre del Municipio"].unique())
```

```python
# Hemos notado que Bogota aparece como Bogota, D.C. por ende los cambiaremos a solo Bogota
df["Nombre del Departamento"] = df["Nombre del Departamento"].replace(r"BOGOTA.*", "BOGOTA", regex=True)
df["Nombre del Municipio"] = df["Nombre del Municipio"].replace(r"BOGOTA.*", "BOGOTA", regex=True)
```

```python
# He notado que en el tipo de Estacion hay valores NaN, por ende los reemplazamos por desconocido
df['Tipo de Estacion'] = df['Tipo de Estacion'].fillna('Desconocido')
# Verificamos nuestros valores
df['Tipo de Estacion'].unique()
```

```python
# Corregimos el tipo de columna de nuestras variables objeto a categoricas
df = df.astype({
    "Variable": "category", "Unidades": "category", "Nombre del Departamento": "category", 
    "Nombre del Municipio": "category", "Tipo de Estacion": "category"
})
```

```python
# Revisamos la informacion de las columnas, que no tengamos datos faltantes 
# y que los tipos de datos sean acordes
import seaborn as sns
df.info()
sns.heatmap(df.isnull(), cbar = False, yticklabels = False)
```

```python
# Revisar cantidad de valores nulos y que porcentaje corresponde
print('Cantidad de valores nulos en la columna "Representatividad Temporal": ', df["Representatividad Temporal"].isnull().sum())
print('Porcentaje de valores nulos en la columna "Representatividad Temporal": ', round(df["Representatividad Temporal"].isnull().mean() * 100,2))
```

Como los valores nulos representan solo el 0.53% de toda la data, eliminarlos no afectará significativamente el análisis. Además, dado que la representatividad temporal indica qué tan confiables y completos son los datos, eliminar las filas con valores nulos ayudará a trabajar con información más precisa.

```python
# Eliminamos los valores nulos de la columna Representatividad Temporal
df = df.dropna(subset=["Representatividad Temporal"])
df.info()
```

```python
# Exportamos los datos limpios en un CSV
df.to_csv('Calidad_Aire.csv', index=False, encoding='utf-8')
```
