## Proyecto Calidad del Aire en Colombia (EDA)

### Acerca del conjunto de datos

Esta base de datos es generada por El Instituto de Hidrolog√≠a, Meteorolog√≠a y Estudios Ambientales ‚Äì (IDEAM), el cual son las autoridades ambientales obligadas a realizar el reporte de la informaci√≥n de calidad del aire o nivel de inmisi√≥n en SISAIRE. Nuestro conjunto de datos se compone de la siguiente informaci√≥n:

* ID Estacion: Identificador √∫nico asignado por el Subsistema de Informaci√≥n sobre Calidad del Aire - SISAIRE a cada estaci√≥n de monitoreo.
* Autoridad Ambiental: Autoridad Ambiental encargada de operar el Sistema de Vigilancia de la Calidad del Aire en la jurisdicci√≥n donde se ubica la estaci√≥n de monitoreo.
* Estaci√≥n: Nombre espec√≠fico que se designa a la estaci√≥n de monitoreo de calidad del aire.
* Latitud: Coordenada geogr√°fica que indica la latitud de la ubicaci√≥n de la estaci√≥n (Sistema de referencia espacial MAGNA- SIRGAS)
* Longitud: Coordenada geogr√°fica que indica la longitud de la ubicaci√≥n de la estaci√≥n (Sistema de referencia espacial MAGNA- SIRGAS)
* Variable: Contaminante atmosf√©rico o variable meteorol√≥gica que se mide en la estaci√≥n de monitoreo de la calidad del aire, de acuerdo con las siguientes abreviaturas: 

    Contaminantes atmosf√©ricos: 
    1. PM10: Material particulado menor a 10 micras. 
    2. PM2.5: Material particulado menor a 2.5 micras. 
    3. PST: Part√≠culas Suspendidas Totales.
    4. NO: √ìxido de Nitr√≥geno.
    5. NO2: Di√≥xido de Nitr√≥geno.
    6. SO2: Di√≥xido de Azufre.
    7. SO2_TR: Azufre Total Reducido.
    8. O3: Ozono troposf√©rico.
    9. CO: Mon√≥xido de Carbono.

    Variables meteorol√≥gicas: 
    1. VViento: Velocidad del Viento.
    2. DViento: Direcci√≥n del viento.
    3. Taire: Temperatura del aire.
    4. Taire2: Temperatura del aire a 2 metros.
    5. Taire10: Temperatura del aire a 10 metros.
    6. Pliquida: Precipitaci√≥n Liquida.
    7. P: Presi√≥n atmosf√©rica.
    8. HAire: Humedad del aire.
    9. HAire2: Humedad del aire a 2 m.
    10. HAire10: Humedad del aire a 10 m.
    11. RGlobal: Radiaci√≥n Solar Global.
    12. RUVb: Radiaci√≥n Ultravioleta.

* Unidades: Unidad de medida empleada para expresar el contaminante atmosf√©rico o variable meteorol√≥gica
* Tiempo de exposici√≥n (horas): Per√≠odo de referencia espec√≠fico establecido normativamente (Resoluci√≥n 2254/2017) para evaluar los niveles de inmisi√≥n o concentraciones de un contaminante atmosf√©rico (solo aplica para contaminantes atmosf√©ricos)
* A√±o: A√±o en el que se realiz√≥ la medici√≥n en la respectiva estaci√≥n.
* Promedio: Promedio anual del nivel de inmisi√≥n o concentraci√≥n a condiciones de referencia o est√°ndar (std) del contaminante atmosf√©rico o promedio anual del valor de la variable meteorol√≥gica.
* Suma: Total acumulado anual del nivel de inmisi√≥n o concentraci√≥n a condiciones de referencia o est√°ndar (std) del contaminante atmosf√©rico o total acumulado anual del valor de la variable meteorol√≥gica.
* No. de datos: Cantidad total de datos validos recopilados durante el a√±o de medici√≥n.
* Representatividad Temporal: Porcentaje del tiempo total de un per√≠odo tiene datos v√°lidos registrados
* Excedencias limite actual: N√∫mero total de veces en que el contaminante atmosf√©rico ha superado el nivel m√°ximo permisible vigente establecido para el respectivo tiempo de exposici√≥n durante el a√±o de medici√≥n.
* Porcentaje excedencias limite actual: Proporci√≥n determinada por el n√∫mero de veces en que el contaminante atmosf√©rico ha superado el nivel m√°ximo permisible vigente establecido para el respectivo tiempo de exposici√≥n, en relaci√≥n con el n√∫mero total de datos seg√∫n el tiempo de exposici√≥n, durante el a√±o de medici√≥n.
* Mediana: Mediana anual del nivel de inmisi√≥n o concentraci√≥n a condiciones de referencia o est√°ndar (std) del contaminante atmosf√©rico o mediana anual del valor de la variable meteorol√≥gica
* Percentil 98: Medida estad√≠stica que indica el valor del contaminante atmosf√©rico o de la variable meteorol√≥gica por debajo del cual se encuentra el 98% de los datos en el grupo durante el a√±o de medici√≥n
percentil_98
* M√°ximo: Valor m√°s alto registrado del contaminante atmosf√©rico o de la variable meteorol√≥gica durante el a√±o de medici√≥n.
* Fechas/horas del m√°ximo: Fecha y hora en las que se registr√≥ el valor m√°ximo durante el a√±o de medici√≥n.
* M√≠nimo: Valor m√°s bajo registrado del contaminante atmosf√©rico o de la variable meteorol√≥gica durante el a√±o de medici√≥n.
* Fechas/horas del m√≠nimo: Fecha y hora en las que se registr√≥ el valor m√≠nimo durante el a√±o.
* D√≠as de excedencias: N√∫mero total de d√≠as en que el contaminante atmosf√©rico ha superado el nivel m√°ximo permisible vigente establecido para el respectivo tiempo de exposici√≥n durante el a√±o de medici√≥n.
* C√≥digo del Departamento: C√≥digo DANE asignado al departamento.
* Nombre del Departamento: Nombre del departamento
* C√≥digo del Municipio: C√≥digo del municipio
* Nombre del Municipio: Nombre del municipio
* Tipo de Estaci√≥n: Ubicaci√≥n de la estaci√≥n (Sistema de referencia espacial MAGNA- SIRGAS)
* Ubicacion: ubicacion

```python
# Mantenemos en un nuevo dataset, solo nuestras variables a usar o importantes
df = calidad_aire[['Variable', 'Unidades', 'Tiempo de exposici√≥n (horas)', 'A√±o', 'Promedio', 'Suma', 
                   'No. de datos', 'Representatividad Temporal', 'Excedencias limite actual', 
                   'Porcentaje excedencias limite actual', 'Mediana', 'Percentil 98', 'M√°ximo', 'M√≠nimo', 
                   'D√≠as de excedencias', 'Nombre del Departamento', 'Nombre del Municipio', 
                   'Tipo de Estaci√≥n']].copy()
df.head()
```

```python
# Editamos los nombres de nuestras columnas. 
# De tal manera que no tengan tildes o simbolos que puedan presentar problemas en el futuro

df = df.rename(columns={'Tiempo de exposici√≥n (horas)': 'Tiempo de exposicion (horas)',
                        'A√±o': 'Year',
                        'M√°ximo': 'Maximo', 
                        'M√≠nimo': 'Minimo', 
                        'D√≠as de excedencias': 'Dias de excedencias', 
                        'Tipo de Estaci√≥n': 'Tipo de Estacion'})
df.head()
```

```python
# Verificamos nuestros cambios en el nombre de columna y su tipo de dato
df.info()
```

```python
# Revisamos los valores unicos en nuestras columnas con el fin de revisar que no tengamos simbolos, 
# errores de escritura o espacios que nos afecten en el futuro

print('--- Valores unicos en el Nombre del Departamento ---')
print(df["Nombre del Departamento"].unique())
```

```python
print('--- Valores unicos en el Nombre del Municipio ---')
print(df["Nombre del Municipio"].unique())
```

```python
print('---Valores unicos en el Tipo de Estacion---')
print(df["Tipo de Estacion"].unique())
```

```python
# Eliminamos todas las tildes en nuestros datos

import unicodedata

# Funci√≥n para remover tildes
def remover_tildes(texto):
    return "".join(
        c for c in unicodedata.normalize("NFD", texto) if unicodedata.category(c) != "Mn"
    )

# Aplicar la funci√≥n a las columnas
df["Nombre del Departamento"] = df["Nombre del Departamento"].apply(remover_tildes).str.upper()
df["Nombre del Municipio"] = df["Nombre del Municipio"].apply(remover_tildes).str.upper()
```

```python
# Verificamos nuestros valores en el nombre del departamento
print(df["Nombre del Departamento"].unique())
```

```python
# Verificamos nuestros valores en el nombre del municipio
print(df["Nombre del Municipio"].unique())
```

```python
# Hemos notado que Bogota aparece como Bogota, D.C. por ende los cambiaremos a solo Bogota
df["Nombre del Departamento"] = df["Nombre del Departamento"].replace(r"BOGOTA.*", "BOGOTA", regex=True)
df["Nombre del Municipio"] = df["Nombre del Municipio"].replace(r"BOGOTA.*", "BOGOTA", regex=True)
```

```python
# He notado que en el tipo de Estacion hay valores NaN, por ende los reemplazamos por desconocido
df['Tipo de Estacion'] = df['Tipo de Estacion'].fillna('Desconocido')
# Verificamos nuestros valores
df['Tipo de Estacion'].unique()
```

```python
# Corregimos el tipo de columna de nuestras variables objeto a categoricas
df = df.astype({
    "Variable": "category", "Unidades": "category", "Nombre del Departamento": "category", 
    "Nombre del Municipio": "category", "Tipo de Estacion": "category"
})
```

```python
# Revisamos la informacion de las columnas, que no tengamos datos faltantes 
# y que los tipos de datos sean acordes
import seaborn as sns
df.info()
sns.heatmap(df.isnull(), cbar = False, yticklabels = False)
```

```python
# Revisar cantidad de valores nulos y que porcentaje corresponde
print('Cantidad de valores nulos en la columna "Representatividad Temporal": ', df["Representatividad Temporal"].isnull().sum())
print('Porcentaje de valores nulos en la columna "Representatividad Temporal": ', round(df["Representatividad Temporal"].isnull().mean() * 100,2))
```

Como los valores nulos representan solo el 0.53% de toda la data, eliminarlos no afectar√° significativamente el an√°lisis. Adem√°s, dado que la representatividad temporal indica qu√© tan confiables y completos son los datos, eliminar las filas con valores nulos ayudar√° a trabajar con informaci√≥n m√°s precisa.

```python
# Eliminamos los valores nulos de la columna Representatividad Temporal
df = df.dropna(subset=["Representatividad Temporal"])
df.info()
```

```python
# Exportamos los datos limpios en un CSV
df.to_csv('Calidad_Aire.csv', index=False, encoding='utf-8')
```
